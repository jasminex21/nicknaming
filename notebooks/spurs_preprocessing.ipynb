```python
TWEETS['during_game'] = False
TWEETS['match_report'] = None

for i, game in spurs_schedule.iterrows():
    
    ko, ft = game["KO_datetime"], game["FT_datetime"]
    match_report = game["match_report"].split("/")[3]

    during_game_mask = (TWEETS["date"] >= ko) & (TWEETS["date"] <= ft)

    TWEETS.loc[during_game_mask, 'during_game'] = True
    TWEETS.loc[during_game_mask, 'match_report'] = match_report

    same_day_mask = (TWEETS["date"].dt.date == ko.date()) & ~during_game_mask

    TWEETS.loc[same_day_mask, 'match_report'] = match_report
```

```python
def _convert_minute(minute):
    """Converts e.g. 90+2 to 92"""

    if '+' in minute:
        base_minute, added_time = minute.split('+')
        return int(base_minute) + int(added_time)
        
    else:
        return int(minute)

def _add_ht(minute):

    if int(minute) > 45:
        return int(minute) + 15
    else:
        return int(minute)
```

```python
EVENTS = pd.read_csv("/home/jasmine/PROJECTS/research/data_scrape/ALL_GAME_EVENTS.csv")
EVENTS
```

```python
game_ids = spurs_schedule["report_id"].unique()
game_ids
```

```python
spurs_events = EVENTS[EVENTS["game_id"].isin(game_ids)]
spurs_events
```

```python
spurs_events["minute_eng"] = spurs_events["minute"].apply(_convert_minute)
spurs_events["minute_eng"] = spurs_events["minute_eng"].apply(_add_ht)
spurs_events
```

```python
TWEETS
```

```python
TWEETS["score"] = None

for i, tweet in TWEETS.iterrows(): 

    if tweet["during_game"]: 

        game_events = spurs_events[spurs_events["game_id"] == tweet["match_report"]]
        recent_events = game_events[tweet["mins_after_ko"] >= game_events["minute_eng"]]

        if not recent_events.empty: 

            most_recent_event = recent_events.iloc[-1]
            TWEETS.at[i, "score"] = most_recent_event["score"]

        else: 

            TWEETS.at[i, "score"] = "0:0"
```

```python
TWEETS[TWEETS["during_game"]]
```

```python
spurs_events["home_team"] = [" ".join(s.split(" ")[1:]).split("-")[0] for s in spurs_events["game"]]
spurs_events["away_team"] = [" ".join(s.split(" ")[1:]).split("-")[-1] for s in spurs_events["game"]]
spurs_events
```

```python
TWEETS["home_team"] = None
TWEETS["away_team"] = None

for i, tweet in TWEETS.iterrows(): 

    if tweet["match_report"]:
        game_events = spurs_events[spurs_events["game_id"] == tweet["match_report"]]
        
        home_team = game_events["home_team"].values[0]
        away_team = game_events["away_team"].values[0]
    
        TWEETS.at[i, "home_team"] = home_team
        TWEETS.at[i, "away_team"] = away_team
TWEETS
```

```python
TWEETS["margin"] = None

for i, tweet in TWEETS.iterrows(): 

    if tweet["during_game"]: 

        scores = tweet["score"].split(":")

        if tweet["home_team"] == "Tottenham": 
            TWEETS.at[i, "margin"] = int(scores[0]) - int(scores[1])

        else: 
            TWEETS.at[i, "margin"] = int(scores[1]) - int(scores[0])
TWEETS
```
