```python
#all_standings = {season: {} for season in SEASONS}
standings_cols = ["Standing", "Club", "Points Pts", 'Won W', 'Drawn D', 'Lost L', 'GF', 'GA', 'GD']
```

```python
dropdowns = driver.find_elements(By.CLASS_NAME, "dropDown")
dropdowns
```

```python
for season in ["2022/23"]: 

    print(f"Scraping season {season}\n\n-----")

    # click the season dropdown
    dropdowns[1].click()
    time.sleep(5)
    # click the season
    option = driver.find_element(By.XPATH, f"//li[@data-option-name='{season}']")
    time.sleep(5)
    option.click() 
    time.sleep(5)

    # loop through each week of the seasons
    for wk in range(8, 39): 
        try:
            # click the GW dropdown
            dropdowns[2].click()
            time.sleep(5)
            # click the week
            wk_option = driver.find_element(By.XPATH, f"//li[@data-option-name='{wk}']")
            time.sleep(3)
            wk_option.click()
            time.sleep(8)
            # grab the standings table as df
            html = driver.page_source
            soup = BeautifulSoup(html, "html.parser")
            table = soup.find("table")
            standings = pd.read_html(str(table))[0]
            # getting rid of extra rows
            standings = standings.iloc[0::2]
            standings["Standing"] = list(range(1, 21))
            standings.reset_index(inplace=True)
            standings = standings[standings_cols]
    
            # store
            all_standings[season][wk] = standings
            print(f"Week {wk} done\n")

        except NoSuchElementException:
            print(f"Skipping week {wk}")
            continue

        except ValueError as e: 
            print(e)
            continue

all_standings
```
