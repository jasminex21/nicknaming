```python
# function that takes in a list of Twitter handles and returns their Twitter user id
def get_twitter_id(handles): 

    driver_path = r"/home/jasmine/chromedriver-linux64/chromedriver"
    service = Service(executable_path=driver_path)
    driver = webdriver.Chrome(service=service)
    converter_link = "https://twitvd.com/twitter-id/"
    driver.get(converter_link)
    
    user_ids = []
    
    for handle in handles: 
        # get search bar
        search_input = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located(
                (By.XPATH, '//input[@placeholder="ex: elonmusk" or @placeholder="username"]')
            ))
        # remove all content from search bar, if any
        search_input.clear()
        # enter handle, press enter
        search_input.send_keys(handle)
        search_input.send_keys(Keys.RETURN)
        # get user ID
        try: 
            res = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.ID, "user-id")))
            # get only the id
            user_id = res.text.split(" ")[-1]
    
            user_ids.append(user_id)
            print(f"{handle} -> {user_id}")

        except TimeoutException: 
            print(f"Skipping {handle}")

        time.sleep(2)

    return user_ids  
```

```python
handles = ALL_ACCOUNTS['account_handle']
handle_ids = get_twitter_id(list(handles))
```
